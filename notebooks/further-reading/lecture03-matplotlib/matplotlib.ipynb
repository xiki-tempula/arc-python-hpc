## Basic plotting
<hr style="border: solid 4px green">

Import `numpy` and `matplotlib`'s plotting functionality via the `pyplot` interface.

> *Note*: when using the notebook, plots can be forced to appear in the browser by adding the "inline" option.

````python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
````

````python
# create some data points for y = cos(x) using numpy
xmin = 0
xmax = 10
npts = 50
x = np.linspace (xmin, xmax, npts)
y = np.cos (x)
````

````python
# do some plots
plt.plot (x, y, "b--") # blue, dotted line
plt.plot (x, y, "ro")  # red dots, no line
plt.show ()
````

## Line and marker styles
<hr style="border: solid 4px green">

Many ways to customise a plot, *e.g.*

````python
# set the figure size (in inches) and plot
fig = plt.figure (figsize=(3,3)) 
plt.plot (x, y, "c-")
plt.show ()
````

## Line and marker styles (cont'd)
<hr style="border: solid 4px green">

Many ways to customise a plot, *e.g.*

````python
# markers and their properties can be controlled
# (unfilled markers: '.',+','x','1' to '4','|')
plt.plot (x,y, "+", markersize=10)
plt.show ()
````

````python
# markers and their properties can be controlled
# (filled markers include: 'o', 's','*','d','>','^','v', 'p', 'h')
plt.plot (x, y, "d", markerfacecolor = "None", markeredgecolor = "g", markersize=10)
plt.show ()
````

## Axes and labels
<hr style="border: solid 4px green">

Set the x-axis and y-axis limits, adjust title font properties.

````python
# set axis limits
plt.xlim ((xmax*0.25, xmax*0.75))
plt.ylim ((np.cos(xmin*0.25), np.cos(xmax*0.75)))
plt.plot (x, y, "mo-")
plt.show ()
````

````python
# set title placement and font properties
plt.plot (x, y, "x")
plt.suptitle ("A plot of $cos(x)$", fontsize = 20)

# location of the title can be controled via "loc" ("center", "left", "right")
# and "verticalalignment": "center", "top", "bottom", "baseline"

plt.title ("A Placed Title", loc = "left", verticalalignment = "top")
plt.show ()
````

## Tickmarks and annotations
<hr style="border: solid 4px green">

Adding custom tickmarks and annotations.

````python
# tick marks: take the default, or set explicitly
fig = plt.figure (figsize=(4,3.5))
plt.plot (x, y, "x")
nticks = 4
tickpos = np.linspace (xmin,xmax,nticks)
labels = np.repeat (["tick"], nticks)matplotlib.
plt.xticks (tickpos, labels, rotation="vertical")
plt.show ()
````

## Tickmarks and annotations (cont'd)
<hr style="border: solid 4px green">

Adding custom tickmarks and annotations.

````python
# arrows and annotations
plt.plot(x, y, 'x')
atext = 'annotate this'
arrowtip = (1.5, 0.5)
textloc = (3, 0.75)
plt.annotate (atext, xy=arrowtip, xytext=textloc,
              arrowprops=dict(facecolor='black', shrink=0.01),)
plt.show ()
````

## Subplots
<hr style="border: solid 4px green">

### The idea
<img src="./images/subplotgrid.png"; style="float: right; width: 40%;">

* a figure can contain multiple plots, or subplots
* use `subplot()` to place plots on a regular grid

```python
subplot(nrows, ncols,
        plot_number)
```

* control which subplot is being used
  * "current" axes is last created
  * or use `pyplot.sca(ax)`

### Subplots and axes objects
Can move between subplots by keeping a reference to the `axes` array.

````python
(fig, axes) = plt.subplots (nrows = 2, ncols = 2)

axes[0,0].plot (x, y, 'b-')
axes[1,1].plot (x, y, 'r-')
plt.show ()
````

Space between subplots may be adjusted.

```python
subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
```

````python
(fig, axes) = plt.subplots (nrows = 2, ncols = 2)
plt.subplots_adjust (wspace = 1.0)

plt.sca (axes[0,1])
plt.plot (x, y, 'b-')
plt.sca (axes[1,0])
plt.plot (x, y, 'r-')
plt.show ()
````

## General subplots using subplot2grid 
<hr style="border: solid 4px green">

<img src="./images/subplot2grid.png"; style="float: right; width: 16%;">

* for more control over subplot layout, use `subplot2grid()`

```python
subplot2grid(shape,
             location,
             rowspan = 1,
             colspan = 1)
```

* subplots can span more than one row or column

````python
fig = plt.figure ()

ax1 = plt.subplot2grid ((3, 3), (0, 0))
ax2 = plt.subplot2grid ((3, 3), (0, 1), colspan=2)
ax3 = plt.subplot2grid ((3, 3), (1, 0), colspan=2, rowspan=2)
ax4 = plt.subplot2grid ((3, 3), (1, 2), rowspan=2)
ax1.plot (x, y, 'r-')
ax2.plot (x, y, 'g-')
ax3.plot (x, y, 'b-')
ax4.plot (x, y, 'c-')
plt.show ()
````

## Solution : Three plots
<hr style="border: solid 4px green">

````python
# %load threeplots.py

import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 1)
fig.set_size_inches(4.0, 8.0)

plt.suptitle("THREE PLOTS", fontsize = 20)
plt.subplots_adjust(left = 0.2)

pie_labels = 'A', 'B', 'C', 'D'
pie_sizes = [15, 30, 35, 20]
pie_colours = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
pie_radius = 1.1
pie_angle = 90

plt.sca(axes[0])
plt.axis('equal')
plt.pie(pie_sizes, labels=pie_labels, colors=pie_colours,
        autopct = '%1.0f%%', shadow = False, startangle = pie_angle,
        radius = pie_radius)

data1 = np.genfromtxt("uniform.dat")

plt.sca(axes[1])
plt.hist(data1, bins = 10, normed = 1, color = "m", label = "Uniform")
plt.ylabel("$P(x)$", size = 18)
plt.legend()

data2 = np.genfromtxt("normal.dat")

plt.sca(axes[2])
plt.hist(data2, bins = 10, normed = True, color = "g", label = "Normal")
plt.xlabel("Variable $x$", size = 18)
plt.ylabel("$P(x)$", size = 18)
plt.legend()

plt.show()
````

## Advanced topic : <span style="font-family: Courier New, Courier, monospace;">matplotlib</span> frontend and backend
<hr style="border: solid 4px green">

`matplotlib` consists of two parts, a frontend and a backend:
 
* *frontend* = the user-facing code, *i.e.* the interface
* *backend* == the engine that performs all the hard work behind-the-scenes to render the image
<br><br>

Two types of backend
* user interface (interactive) backends
* hardcopy (non-interactive) backends to make image files
  * *e.g.* Agg (png), Cairo (svg), PDF (pdf), PS (eps, ps)
* can be checked with `matplotlib.get_backend()`
* switch to a different backend (*before* importing `pyplot`) with

```python
matplotlib.use(...)
import matplot.pyplot as plt
...
```

> More information: http://matplotlib.org/faq/usage_faq.html#what-is-a-backend

## Advanced topic : interactive mode and the iPython shell
<hr style="border: solid 4px green">

* By default, the iPython shell will not display figures to the screen until the `plt.show()` command is issued
  * this command "blocks" any further commands being issued in the shell
  * close the figure using the window manager to continue issuing commands in the shell

* In "interactive" mode, the iPython backend points to an "interactive" backend, which allows a figure display to screen as soon as either `plt.figure()` or `plt.plot()` is called
  * the displayed figure can be updated "interactively" by issuing further commands in the shell

* To run an iPython shell in "interactive" mode, can launch the shell with
```bash
bash> ipython --matplotlib
```

* Alternatively, issue `%matplotlib` from within the shell (this must be done before importing anything from `matplotlib`)

* An "interactive" backend does not however guarantee your figures will automatically display to screen (it's not as simple as that, see link below)

* You can query if your shell is in "interactive" mode with: `matplotlib.is_interactive()`
<br><br>

> For more information: http://matplotlib.org/faq/usage_faq.html#what-is-interactive-mode
